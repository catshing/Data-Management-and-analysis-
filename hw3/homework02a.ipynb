### Origin of the data: 

I downloaded the data from Global Terrorism Database(GTD), which is an open-source database which has records of terrorist events. It is one of the most comprehensive unclassified database of terrorist events around the world. The csv file that I downloaded has over 180,000 records of terroris incidents and attacks from all over the world internationally and domestically in the US as well. The information contained in the file is based on reports from a wide range of open media sources (over 4,000,000 news articles and 25,000 news articles were reviewed to collect the data) and the GTB team only added records only if they are credible and verified. Due to the transfer of the data in 1993, a lot of data in 1993 was missing but the GPB team decided to not include the data in 1993.

The Pinkerton Global Intelligence Services(PGIS) identified and recorded terrorism incidents from governemnt reports, wire services, and top international newspaper to determine the level of terrorism from 1970 to 1997. The Global Terrorism Database (GTD) started in 2001 when researched at University of Maryland took over the database from PGIS. The Department of Homeland Security provided additional funding to GTD from 1997.

GTD team utilises a wide range of news media sources for the identification of the incidents by using machine learning and and data mining techniques to identify types of terrorist incidents in news articles and also developed a Data Management System to compile the database.

##### Transformating and cleaning the data 
1. Drop columns that are irelevant or unncessary to what I want to find from the data 
2. Rename columns so that it is easier to understand  
3. Make the country column all capital letters so that it is easier read since I will be dealing 
4. Fill all NaN to empty strings 
5. Use lambda functions to convert iyear column to integet datatype 
6. Sort values by year, month and day 

```python
df = df.rename(columns = {'iyear': 'year', 
                          'imonth': 'month',
                          'iday': 'day', 
                          'country_txt': 'country',
                          'region_txt': 'region',
                          'provstate': 'state',
                          'nkill': 'no_kills'
                         })
    
df['country'] = df['country'].str.upper()
kills = df['no_kills']
kills.fillna(0, inplace = True)
df = df.fillna('')
to_int = [c for c in df.columns if "iyear" in c] 
    
for c in to_int:
    df[c] = df[c].apply(lambda x: pd.astype(x))
        
df.sort_values(by=['year','month', 'day'], inplace=True) 
```

In order to find out the success rate of the terrorist attacks from 1970 to 2017, I use the groupby operation to group the day and success together by year. The 2 columns: day and success are total number of days of terrorist attacks and the number of days that are successful. Therefore, I can divide success by the number of days of terrorists attacks to find the success rate of terrorist attacks per year

```python
successRate = df.groupby(['year'])[['day','success']].sum()
successRate
```

```python
successRate['success'] = list(map(lambda x,y: x/y, successRate['success'],successRate['day']))
```

```python
successRate.drop(['day'], axis = 1, inplace = True)
```

The plot shows the success rate of terrorist attacks from 1970 to 2017 in a line graph. It shows the the success rate is reasonably low in the early 1970s and gradually increased from late 1970s and have a constant success rate of over 0.05 till 2000. However, from the early 2000s, there was a dramatic increase in success rate which can be due to factors such as war (Iraq War). However, success rate started to decrease a lot starting from late 2000s to 2010s. Reasons could be technology has advanced and it became easier to prevent terorrists attacks. 

```python
successRate.plot(title='Success rate of terrorist attacks from 1970 to 2017', kind='line', color='red',figsize=[19,10])
plt.xlabel('Year', size=10)
plt.ylabel('Success Rate', size=10)
```
