Find the top 5 frequent words in Anne of Green Gables

1. Use error handling (try and except) to check if file exists
2. Use from heapq import nlargest to get the top n elements 
3. Create an empty dictionary 
4. Replace the following punctuation with an empty string
5. In order to ignore empty spaces, use split function to get each word seperately 
6. If the word is in the dictionary, then increment the value of that word by 1
7. If not, then asssign the value of that word to 1
8. In the end, use the nlargest function to get the top 5 frequent words

```python
try: 
    with open('green_gables.txt') as f:
        lines = f.read()
        print (lines[0])
except FileNotFoundError:
    print('sorry done')
else: 
    print('lines[0]')
print('done')
```

```python
from heapq import nlargest
dict = {}
for each in '()-_=+,<.>/?;:[{]}~`\|"''!@#$%^&*''\'':
    lines = lines.replace(each,"")
words_split = lines.split(' ')
for word in words_split:
    if word in dict:
        dict[word] +=1
    else:
        dict[word] = 1
print(dict)
```

```python
top_five = nlargest(5, dict, key=dict.get)
print(top_five)
```
